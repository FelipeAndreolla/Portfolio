Importando as bibliotecas pandas, numpy e matplotlib.pyplot, que são usadas para manipular dados, realizar operações matemáticas e criar gráficos, respectivamente.

Essas são algumas bibliotecas Python muito populares e úteis para computação científica e análise de dados. Vou explicar brevemente o que cada uma faz:

- **pandas** é uma biblioteca para manipular e analisar dados estruturados e tabulares. Ela oferece estruturas de dados como Series e DataFrame, que facilitam a leitura, filtragem, agrupamento e visualização de dados de diferentes fontes.

- **matplotlib** é uma biblioteca para criar gráficos estáticos, animados e interativos em Python. Ela tem uma interface semelhante ao MATLAB e permite personalizar o estilo e o layout dos gráficos. Ela também suporta vários formatos de arquivo e ambientes interativos.
- **scipy** é uma biblioteca que fornece algoritmos fundamentais para computação científica em Python. Ela inclui módulos para otimização, integração, interpolação, problemas de autovalores, equações algébricas, equações diferenciais, estatística e muitas outras classes de problemas.

Se a coluna "Unnamed: 0" não contém informações úteis e parece ser um índice numérico adicional, posso optar por removê-la do DataFrame.

Para excluir a coluna "Unnamed: 0", vou usar o seguinte código:

````python
# Removendo colunas Unnamed

df = df.drop("Unnamed: 0", axis=1)
````

Explicando o código em detalhes:

```python
df = df.drop("Unnamed: 0", axis=1)
```

- A função `drop()` é usada para remover uma coluna (ou linha) de um DataFrame.
- No código acima, estamos chamando a função `drop()` no DataFrame `df`.
- O primeiro argumento da função `drop()` é o nome da coluna que queremos remover, que é "Unnamed: 0" no caso.
- O segundo argumento, `axis=1`, indica que queremos remover uma coluna (o eixo 1 representa as colunas). Se você quisesse remover uma linha, usaria `axis=0`.
- O resultado da função `drop()` é atribuído novamente à variável `df`, atualizando o DataFrame sem a coluna "Unnamed: 0".

Ao executar esse código, a coluna "Unnamed: 0" será removida do DataFrame `df`. Podemos usar a função `df.head()` novamente para verificar se a coluna foi excluída com sucesso.

É importante observar que, se você tiver outras colunas que deseja remover ou se deseja preservar a coluna "Unnamed: 0", basta adaptar o código conforme necessário.

````python
# Converter a coluna "release_date" em formato de data
df['release_date'] = pd.to_datetime(df['release_date'])

# Tratar valores ausentes na coluna "year"
df['year'] = df['release_date'].dt.year
df['year'] = df['year'].fillna(-1)  # Substituir valores ausentes por um valor sentinela, como -1

# Converter a coluna "year" para inteiro
df['year'] = df['year'].astype(int)

# Extrair o mês em uma nova coluna "month" como objeto
df['month'] = df['release_date'].dt.strftime('%B')
````

Lidar com os valores ausentes no nosso conjunto de dados. Para isso, vamos usar o método `isnull().sum()` para verificar a contagem de valores ausentes em cada coluna.

Aqui está o código para realizar essa etapa:

```python
# Verificar a contagem de valores ausentes em cada coluna
missing_values = df.isnull().sum()

# Exibir a contagem de valores ausentes
print(missing_values)
```

Ao executar esse código, você obterá a contagem de valores ausentes em cada coluna do DataFrame. Com base nessa informação, você pode decidir como tratar os valores ausentes de acordo com a natureza dos dados.

Se encontrar valores ausentes, aqui estão algumas abordagens comuns para tratá-los:

- Preencher os valores ausentes com a média ou mediana para dados numéricos, usando o método `fillna()`.
- Preencher os valores ausentes com a moda para dados categóricos, usando o método `fillna()`.
- Remover as linhas com valores ausentes, usando o método `dropna()`.
- Remover as colunas com valores ausentes, usando o método `drop()`.

````python
# Remover as linhas com valores ausentes na coluna "original_title"
df = df.dropna(subset=['original_title'])
````

````python
missing_values = df.isnull().sum()

# Exibir a contagem de valores ausentes
print(missing_values)
````

Explorando as variáveis categóricas do conjunto de dados. Para isso, usaremos o método `value_counts()` para identificar as categorias exclusivas e suas contagens em cada coluna categórica.

Ao executar o código, vou obter a distribuição das categorias exclusivas e suas contagens para cada coluna categórica no DataFrame. Isso fornecerá insights valiosos sobre a distribuição dos dados e o impacto das variáveis categóricas nas análises futuras.


````python
# Identificar as colunas com variáveis categóricas
categorical_columns = df.select_dtypes(include=['object']).columns

# Examinar as categorias exclusivas e suas contagens para cada coluna categórica
for column in categorical_columns:
    print(f"Coluna: {column}")
    print(df[column].value_counts())
    print()
````

````python
# Estatísticas descritivas das colunas numéricas
numeric_columns = ['popularity', 'vote_average', 'vote_count']
numeric_stats = df[numeric_columns].describe()
print(numeric_stats)
````

````python
import matplotlib.pyplot as plt

# Selecionar os 10 filmes com melhor popularidade
top_10_popularidade = df.nlargest(10, 'popularity')

# Extrair os dados relevantes
titulos = top_10_popularidade['original_title']
popularidade = top_10_popularidade['popularity']

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(titulos, popularidade)
plt.title('Top 10 Filmes com Melhor Popularidade')
plt.xlabel('Título do Filme')
plt.ylabel('Popularidade')
plt.xticks(rotation=90)
plt.show()
````

````python
import matplotlib.pyplot as plt

# Ordenar o DataFrame pela coluna 'vote_average' em ordem crescente
df_sorted = df.sort_values('vote_average')

# Criar o gráfico de barras
plt.figure(figsize=(10, 4))
plt.bar(df_sorted['original_title'], df_sorted['vote_average'])
plt.xticks(rotation=90)
plt.xlabel('Título')
plt.ylabel('Vote Average')
plt.title('Classificação média dos Filmes (Ordem Crescente)')
plt.show()
````
